```python
from os import truncate
#%%writefile browser_manager_script.py

import asyncio
from playwright.async_api import async_playwright
from google.colab import files

class BrowserManager:
    def __init__(self):
        self.browser = None
        self.page = None
        self.pages = []



    async def start_browser(self):
        """this method is to launch the browser"""
        playwright = await async_playwright().start()
        self.browser = await playwright.chromium.launch(headless=True)
        self.page = await self.browser.new_page()
        print("Browser launched successfully!")


    async def close_browser(self):
        """this method closes the browser"""
        if self.browser:
            await self.browser.close()
            self.browser = None
            self.page = None
            self.pages = []
            print("Browser closed successfully!")
        else:
            print("Browser not closed yet!")


    async def get_page(self):
        """this method returns the page"""
        return self.page


    async def get_browser(self):
        """this method returns the browser!"""
        return self.browser


    async def navigate_to_page(self, url: str = None):
        """Navigation to a specific oage with a known url"""
        await self.page.set_extra_http_headers({
                      "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
                              })

        await self.page.goto(url, wait_until= "domcontentloaded")
        #Lets take a screenshot of the damn page as evidence,hahaüòÇ
        shot_path = f"Evidnece_{url.replace('https://', '').replace('/', '_').replace('.', '_')}.webp"
        await self.page.screenshot(path=shot_path)
        files.download(shot_path)
        print(f"Navigated to the page {url} successfully and saved a copy as evidence or proof of visitation!")


    async def manage_tabs(self, url):
        """Open a new tab and navigate to a url"""
        #self.browser = await playwright.chromium.lauch(headless= False)
        if not self.browser:
            print("Browser is not initialised")
            return
        print("Opening a new tab....")
        if url is None:
            raise ValueError("URL required to open a new tab!")
        self.new_page = await self.browser.new_page()
        await self.new_page.goto(url)
        print(f"Navigated to a new tab for {url} successfully!")
        # Take a screenshot of the new tab
        screenshot_path = f"screenshot_{url.replace('https://', '').replace('/', '_').replace('.', '_')}.png"
        await self.new_page.screenshot(path=screenshot_path)
        files.download(screenshot_path)
        print(f"Screenshot of the new tab saved as {screenshot_path}")


    async def screenshot_page(self):
        """Take s screenshot of the current page"""
        #path = "screenshot.png"
        await self.page.screenshot(path= "screenshot.jpg")
        evidence = files.download("screenshot.jpg")
        if evidence == True:
            print("Screenshot taken successfully!")
        else:
            print("Visited but not downloaded!")


    async def emulate_device(self, device_name):
        """Simulate a specific device"""
        from playwright.async_api import devices
        device = devices[device_name]
        context = await self.browser.new_context(**device)
        self.page = await context.new_page()
        print(f"Emulated device {device_name} successfully!")


    async def automate_login(self, url,username_selector: str, password_selector: str, submit_selector: str, username: str, password: str,):
        """Automating login process using playwright
        if not self.page:
            print("page is not opend yet")
            return"""
        try:
            await self.page.goto(url)
            await self.page.wait_for_selector(username_selector, timeout= 50000)
            await self.page.fill(username_selector, username)
            await self.page.wait_for_selector(password_selector, timeout= 50000)
            await self.page.fill(password_selector, password)
            await self.page.wait_for_selector(submit_selector, timeout= 50000)
            await self.page.click(submit_selector)
            print("Logged in successfully!")
        except Exception as e:
            #content = await self.page.content()
            print(f"Current page error: {e}")


#We would create the form filling automator in a separate file for complications purposes

    """async def fill_and_submit_form(self, url, form_selector: str, form_data: dict, submit_selector: str):
        Fill and subkit a web form automatically with dynamic details
        if not self.page:
            print("Page mot opened yet")
            return
        await self.page.goto(url)
        await self.page.wait_for_selector(form_selector)
        form = self.page.locator(form_selector)
        await form.wait_for()
        for selector, value in form_data.items():
            await form.fill(selector, value if value is not None else "") #This is going to fill a field with an empty string if value is not specified
        await form.click(submit_selector)
        print("Form filled and submitted successfully!")"""


    async def mouse_and_keyboard_interaction(self, actions: list):
        """Performing advanced mouse and keyboard testing interactions"""
        if not self.page:
            print("page not yet opened")
            return
        for action in actions:
            if action["type"] == "click":
                await self.page.click(action["selector"])
                print("Clicked on an element successfully!")
            elif action["type"] == "hover":
                await self.page.hover(action["selector"])
                print("Hovered over an element successfully!")
            elif action["type"] == "type":
                await self.page.type(action["selector"], action["value"])
                print("Typed into an element successfully!")
            elif action["type"] == "press":
                await self.page.press(action["selector"], action["key"])
                print("Pressed a key on an element successfully!")
            else:
                print("Invalid action type!")
        print("Mouse and keyboard interaction performed successfully!")



if __name__ == "__main__":
    display = Display(visible=1, size=(1920, 1080), backend= "xvfb")
    display.start()
    async def test_func():
        try:
            automated = BrowserManager()
            await automated.start_browser()
            #await automated.navigate_to_page("https://portal.altschoolafrica.com/auth/signin?callbackUrl=%2F")
            await automated.automate_login(url= "https://www.saucedemo.com/", username_selector= "#user-name", password_selector= "#password", submit_selector= "#login-button", username = "standard_user", password= "secret_sauce")
            await automated.close_browser()
        except Exception as e:
            print(f"an error occurred: {e}")
await test_func()
```
